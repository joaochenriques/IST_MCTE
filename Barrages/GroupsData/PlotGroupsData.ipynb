```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mpl
%config InlineBackend.figure_formats = ['svg']
mpl.rcParams["figure.figsize"] = (7, 3)
```

```python
df = pd.read_excel( 'Data.xlsx', sheet_name='Sheet1', header=0, skiprows=(1,) )
df
```

```python
def plot_time_series( N ):
  ζ1 = df['𝜁1'][N]
  ζ2 = df['𝜁2'][N]
  ω1 = df['𝜔1'][N]/3600.0
  ω2 = df['𝜔2'][N]/3600.0
  φ1 = df['𝜙1'][N]
  φ2 = df['𝜙2'][N]

  name = df['Name'][N]
  tide_period = 1.0 / ( ω2 - ω1 ) * 2.0*np.pi

  t = np.linspace( 0, tide_period, 1000 )
  A = ζ1 * np.cos( ω1 * t + φ1 ) + ζ2 * np.cos( ω2 * t + φ2 )

  t_hours = t / 3600.0
  mpl.plot( t_hours, A )
  mpl.title( name );
```

```python
plot_time_series( 0 )
```

```python
plot_time_series( 1 )
```

```python
plot_time_series( 2 )
```
