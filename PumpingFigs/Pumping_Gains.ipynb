```python
import numpy as np
import matplotlib.pyplot as mpl
import mpl_utils as mut
%config InlineBackend.figure_formats = ['svg']
mut.config_plot()
```

```python
def gains( ηp, H, ΔH ):
    sq = ( H + ΔH )**2
    return ηp * sq / ( sq - H**2 ) - 1
```

```python
H = 2
ΔH = 1

rem = '$H=%.0f\,$m, $\Delta H=%.0f\,$m' % ( H, ΔH )

ηt_vals = np.linspace(0.6, 1, 5)
ηp_vals = np.linspace(0.5, 1.0, 30)

g_vals = gains( ηp_vals, H, ΔH )

for ηt in ηt_vals: 
  mpl.plot( ηp_vals, ηt*g_vals, label = ( "$\eta_\mathrm{turb}=%.2f$" % ηt ) )

mpl.gca().set_axisbelow(True)
mpl.grid()
mpl.legend(loc="upper left")
mpl.xlabel("$\eta_\mathrm{pump}$")
mpl.ylabel("$E_\mathrm{turb}/E_\mathrm{pump}-1$")

mpl.text( 0.865, 2.25, rem, bbox=dict(facecolor='w', edgecolor='k', boxstyle="round", alpha=1) )
mpl.text( 0.435, 2.5, 'a)', fontsize=18 )
mpl.ylim( -0.1, 2.5 )
mpl.savefig( 'Pumping_case1.pdf', bbox_inches='tight' )
```
