```python
import numpy as np
import matplotlib.pyplot as mpl
import mpl_utils as mut
%config InlineBackend.figure_formats = ['svg']
mut.config_plot()

from matplotlib.patches import Rectangle
```

```python
ηt_vals = np.linspace(0.4, 1.0001,  7)
ηp_vals = np.linspace(0.3, 1.0001, 30)

mpl.gca().axhline(1, color = '#FF8080', linewidth=2 )

for ηt in ηt_vals: 
  g_vals = 3.0 - 1.0 / ( ηt * ηp_vals )
  mpl.plot( ηp_vals, g_vals, label = ( "$\eta_\mathrm{turb}=%.2f$" % ηt ) )

mpl.gca().add_patch( Rectangle((0, -0.1), 1.0, 1.1, facecolor = '#FFAAAA', alpha=0.3 ) )

mpl.gca().set_axisbelow(True)
mpl.grid()
mpl.legend(loc="upper left", fontsize=12)
mpl.xlabel("$\eta_\mathrm{pump}$")
mpl.ylabel("$\mathrm{Gain}_\mathrm{pumping}$")

mpl.xlim(  0.3, 1.0 )
mpl.ylim( -0.1, 2.2 )

# mpl.plot( 1.0/np.sqrt(2), 1.0, 'o', markersize=7, markeredgecolor='k', 
#           markerfacecolor='w', markeredgewidth=1 )

mpl.savefig( 'Pumping_gains.pdf', bbox_inches='tight' )
```
